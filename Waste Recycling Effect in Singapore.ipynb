# Data Analysis & Visualization

```python
grouped = df1.groupby('year')['waste_type'].max()

# Plot the data
plt.plot(grouped.index, grouped.values, label='Waste Generated')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Waste Generated (kg)')
plt.title('Total Type of Waste Generated Each Year')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

This graphs shows the maximum amount of waste that was generated from 2000 to 2015. This clearly depicts that the city/town/county has been producing same amount of wood waste over the past 15 years and management has been doing nothing to prevent it. Because this shows there is a uniform release of wood waste and this could lead to deforestration. Concerned authorities can take help from our insight and give awareness regarding wood waste.

```python
grouped = df1.groupby('year')['waste_type'].min()
print(grouped)
# Plot the data
plt.plot(grouped.index, grouped.values, label='Waste Generated')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Waste Generated (kg)')
plt.title('Total Type of Waste Generated Each Year')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

This graph shows the minimum waste that has been generated over the past 14 years but there is no behavioral change in it. It has been continously generated at unform rate and there is no significant decrease in it. Ash waste is a by-product of burning coal, wood, or other fuels and contains harmful pollutants like heavy metals and dioxins. If not properly disposed of, ash waste can leach into groundwater and soil, contaminating them and causing harm to the environment.

Sludge waste is the semi-solid residue generated from the treatment of wastewater. It contains high levels of pollutants, including pathogens, heavy metals, and organic matter. If not properly managed, sludge waste can pose a significant risk to human health and the environment. Improper disposal of sludge waste can result in the release of harmful pollutants into the air and groundwater, leading to contamination and potential harm to human health and the environment.

Therefore, it is crucial to properly manage ash and sludge waste to minimize the harm it can cause to the environment and human health. This includes proper disposal methods, such as landfilling, incineration, or treatment and reuse as fertilizer or energy. Proper waste management and recycling programs can help reduce the harm caused by ash and sludge waste and promote social and environmental improvement.

Concerned authorities can take action by taking advantage of this insight and despite being minimum in generation in environment it can still be a alarming risk in future.

```python
grouped = df1.groupby('waste_type')['recycling_rate'].max().reset_index()
# Plot the data
sns.barplot(x='waste_type', y='recycling_rate',hue = "recycling_rate" ,data=grouped)

# Add labels and title
plt.xticks(rotation =90)
plt.xlabel('Waste Type')
plt.ylabel('Recycling Percentage')
plt.title('Maximum Recycling Percentage by Waste Type')

# Show the plot
plt.show()
```

By visualizing the type of waste that is recycled the most, decision-makers can prioritize their efforts and allocate resources towards recycling that specific type of waste. This can lead to an increase in overall recycling rates and a reduction in the amount of waste that ends up in landfills. Furthermore, reducing the amount of waste in landfills can help to minimize the negative impact on the environment, such as soil and water pollution, greenhouse gas emissions, and land degradation. Additionally, by encouraging people to recycle more, there can be a positive impact on the community, creating jobs and reducing the need for new landfills. Overall, this visualization can be a valuable tool for improving the social and environmental conditions in the study area.

```python
grouped = df1.groupby('year')['recycling_rate'].mean().reset_index()

# Plot the data
plt.plot(grouped['year'], grouped['recycling_rate'])

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Recycling Percentage')
plt.title('Trend in Average Recycling Percentage Over the Years')

# Show the plot
plt.show()
```

This graph shows hwo the recycling rate has been fluctuating in the past 15 years. We could clearly see that from year 2000 to 2002 recycling rate was significantly reduced and then from 2004 to 2006 it becomes almost constant for different waste materials but in year 2008 to 2010 it was significantly reduced this might be due to government policies and priorities, budget and natural disaster. But the positive thing is on an average it is continously incrasing from all these years which is encouraging and it means that authorities have been successful in giving awareness to society regarding waste management and recycling.

```python
grouped = df1.groupby('waste_type')['recycling_rate'].mean().reset_index()
plt.bar(grouped['waste_type'], grouped['recycling_rate'])

# Add labels and title
plt.xticks(rotation = 90)
plt.xlabel('Waste Type')
plt.ylabel('Recycling Percentage')
plt.title('Average Recycling Percentage by Waste Type')

# Show the plot
plt.show()
```

This graph shows on an average which type of waste has the highest recycling rate. This insight could again help governement to use their expensive resources in the right direction for the right type of waste. Looks like authorities have been spending much on construction waste because these years have proved to be very crucial for infrastructure development all over the world. But we cannot defy the problems of deforestration, environmental pollution etc. SO this insight could be of great use.

```python
#detecting outliers
sns.boxplot(x='recycling_rate', data=df1)

# Add labels and title
plt.xlabel('Recycling Percentage')
plt.title('Recycling Percentage Outliers')

# Show the plot
plt.show()
```

The presence of outliers in waste generation or recycling percentage data can provide valuable insights into the environmental impact of waste and recycling practices. For example, outliers can indicate areas with particularly high or low levels of waste generation or recycling, which can provide important information for policymakers and environmental organizations to target their efforts.

For example, if a particular year or region has a high recycling rate compared to others, it can be further studied to understand the factors contributing to this success, such as effective recycling programs, strong community engagement, or favorable government policies. On the other hand, if a year or region has a low recycling rate, it can be investigated to determine what factors are hindering the success of recycling programs, such as lack of infrastructure, low awareness, or economic barriers.

This was the purpose of this graph but if we observe the IQR (Inter quartile range) which is the upper and lower quartile no significant outliers exists. This could be possible that the data that has been collected is not enough or there are no extreme regions or years in which recycling rate was maximum. The distribution of recyling rate over the past 15 years has been not normal and it represents the skewyed behavior.

```python
grouped = df1.groupby('waste_type')['recycling_rate'].mean().reset_index()
plt.figure(figsize=(10, 10))
# Plot the data as a pie chart
plt.pie(grouped['recycling_rate'], labels=grouped['waste_type'], autopct='%1.1f%%')
# Add title
plt.title('Waste Type and Recycling Rate')

# Show the plot
plt.show()
```
